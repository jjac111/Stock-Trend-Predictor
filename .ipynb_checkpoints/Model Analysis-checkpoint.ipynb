```python
import tensorflow as tf
import keras
import pandas as pd
from matplotlib import pyplot as plt
import os
from create_model import create_model, lr
from keras.optimizers import Adam
import pickle
from IPython.core import display as ICD
```

```python
# The time ranges that the trained models will predict
time_range = [1, 5, 10, 20, 90, 270]
time_words = {1:'day', 5:'week', 10:'two_weeks', 20:'month', 90:'four_months', 270:'year'}

# The time ranges used for feature calculations
periods = [5, 10, 20, 90, 270]

model_dir = 'models_3/'
history_dir = model_dir + 'history/'
data_dir = 'data_2/final/'
tickers_file = 'ndxt_tickers.txt'
ndxt_tickers = []
with open('data_2/'+tickers_file) as f:
    for ticker in f:
        ndxt_tickers.append(ticker.replace('\n', ''))
```

```python
for t in time_range:
    
    print(f'Models that predict {time_words[t]} into the future:')
    plt.figure(figsize=(16,16))
    
    for i, n1 in enumerate(periods):
        for j, n2 in enumerate(periods):
            
            idx = i + 5*j + 1
            
            plt.subplot(5, 5, idx)

            name = f'{t}-{n1}-{n2}'
            
            filepath = history_dir + name + '_history.csv'

            df = pd.read_csv(filepath)

            plt.plot(df['accuracy'])
            plt.title(name)
            plt.ylim((0.5,1))
            
            plt.grid(True)
            
    plt.tight_layout(pad=3.0)
    plt.show()
```

## Evaluation: whole testing set

```python
# Evaluate each trained model on thei respective testing set.

evaluation = {}
j = 0
for t in time_range:
    for n1 in periods:
        for n2 in periods:
            
            num_models = len(time_range) * len(periods) * len(periods)
            name = f'{t}-{n1}-{n2}'
            filepath = model_dir + name + '.h5'
            model = create_model(t, n1, n2)
            model.load_weights(filepath)
            model.compile(optimizer=Adam(learning_rate=lr), loss='binary_crossentropy', metrics=['accuracy'])
    
            evaluation[name] = {}
            evaluation[name]['evaluations'] = {}

            for symbol in ndxt_tickers:
                evaluation[name]['evaluations'][symbol] = []
                data_X = pd.read_csv(f'{data_dir}{time_words[t]}/{n1}_{n2}/test/{symbol}.csv', header=None).drop(labels=0, axis=1)
                data_y = pd.read_csv(f'{data_dir}{time_words[t]}/{n1}_{n2}/test/{symbol}.csv', header=None)[0]

                samples = len(data_X)
                evaluation[name]['samples'] = samples
                for i in range(samples):
                    x = data_X[i:i+1]
                    y = data_y[i:i+1]
                    evaluation[name]['evaluations'][symbol].append(model.evaluate(x, y, verbose=0)[1])
            j += 1
            print(f'{round(j*100/num_models, 1)}%', end=' ')
            
with open('results/evaluation.pickle', 'wb') as f:
    pickle.dump(evaluation, f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Calculate average accuracies for each symbol in evaluations of every model
acc_streaks = {}
for mod, evals in evaluation.items():
    acc_streaks[mod] = {}
    acc_streaks[mod]['accuracy'] = []
    acc_streaks[mod]['streak'] = []
    acc_streaks[mod]['samples'] = evals['samples']
    
    
    for symbol, accuracies in evals['evaluations'].items():
        acc = round(sum(accuracies)/len(accuracies),2)
        
        streak = 0
        for i, a in enumerate(accuracies):
            if a != 1: 
                streak = i
                break
        
        acc_streaks[mod]['accuracy'].append(acc)
        acc_streaks[mod]['streak'].append(streak)
        
        
```

```python
plot_1 = {mod:vals for mod, vals in acc_streaks.items() if mod.split('-')[0] == '1'}
plot_5 = {mod:vals for mod, vals in acc_streaks.items() if mod.split('-')[0] == '5'}
plot_10 = {mod:vals for mod, vals in acc_streaks.items() if mod.split('-')[0] == '10'}
plot_20 = {mod:vals for mod, vals in acc_streaks.items() if mod.split('-')[0] == '20'}
plot_90 = {mod:vals for mod, vals in acc_streaks.items() if mod.split('-')[0] == '90'}
plot_270 = {mod:vals for mod, vals in acc_streaks.items() if mod.split('-')[0] == '270'}

to_plot = [plot_1, 
           plot_5, 
           plot_10, 
           plot_20, 
           plot_90, 
           plot_270]

plt.figure(figsize=(18,12))
ylim = max([item for sublist in [v['accuracy'] for mod, v in acc_streaks.items()] for item in sublist])

for i, p in enumerate(to_plot):
    accuracies = []
    streaks = []
    plt.subplot(2, 3, i+1)
    plt.grid()
    
    for mod, vals in p.items():
        avg_acc = sum(vals['accuracy'])/len(vals['accuracy'])
        avg_streak = sum(vals['streak'])/len(vals['streak'])
        accuracies.append(avg_acc)
        streaks.append(avg_streak)
    
    days = mod.split('-')[0]
    mods = list(p.keys())
    samples = vals['samples']
    labels = [m.split('-', 1)[1] for m in mods]
    plt.bar(x= range(len(mods)), height= accuracies)
    plt.xticks(ticks= range(len(mods)), labels=labels, rotation=70)
    plt.ylim((0,ylim))
    plt.xlabel('Models: n1-n2')
    plt.ylabel('Average Accuracy')
    plt.title(f'Models that predict {days} day(s) ahead.\nEvaluated on {samples} samples each symbol, {samples*len(ndxt_tickers)} each model.')

plt.tight_layout(pad=3.0)

plt.show()
```

```python
# Construct a DataFrame for the evaluation data

df = pd.DataFrame(acc_streaks.items(), columns=['model', 'dict',])

df['streak'] = df['dict'].map(lambda x: x['streak']).map(lambda x: round(sum(x)/len(x)))
df['accuracy'] = df['dict'].map(lambda x: x['accuracy']).map(lambda x: round(sum(x)/len(x), 3))
df['samples per symbol'] = df['dict'].map(lambda x: x['samples'])
df.drop(['dict'], inplace=True, axis=1)

df_sorted = df.sort_values('accuracy', ascending=False).head(10)[['model','accuracy', 'samples per symbol', 'streak']]
minim = min(df_sorted['accuracy'])
maxim = max(df_sorted['accuracy'])
print(f'The top 10 accuracies of evaluation, ranging from {round(minim*100)}% to {round(maxim*100)}%')
ICD.display(df_sorted)

for days in time_range:
    days = str(days)
    print(f'Top accuracies for {days} day(s) evaluations.')
    df_ = df[df['model'].map(lambda x: x.split('-')[0] == days)].sort_values('accuracy', ascending=False).head(5)
    df_['t'] = df_['model'].map(lambda x: x.split('-')[0])
    df_['n1'] = df_['model'].map(lambda x: x.split('-')[1])
    df_['n2'] = df_['model'].map(lambda x: x.split('-')[2])
    
    ICD.display(df_[['t', 'n1', 'n2', 'accuracy', 'samples per symbol', 'streak']])
```

```python
# Plot the accuracy of each group of models 

plt.figure(figsize=(18,12))
ylim = max([sum(v['streak'])/len(v['streak']) for mod, v in acc_streaks.items()] )

for i, p in enumerate(to_plot):
    streaks = []
    plt.subplot(2, 3, i+1)
    plt.grid()
    
    for mod, vals in p.items():
        avg_streak = sum(vals['streak'])/len(vals['streak'])
        streaks.append(avg_streak)
    
    days = mod.split('-')[0]
    mods = list(p.keys())
    samples = vals['samples']
    labels = [m.split('-', 1)[1] for m in mods]
    plt.bar(x= range(len(mods)), height= streaks, color='#ff1a34')
    plt.xticks(ticks= range(len(mods)), labels=labels, rotation=70)
    plt.ylim((0,ylim))
    plt.xlabel('Models: n1-n2')
    plt.ylabel('Average Acurate Streak')
    plt.title(f'Models that predict {days} day(s) ahead.\nEvaluated on {samples} samples each symbol, {samples*len(ndxt_tickers)} each model.')
    

plt.tight_layout(pad=2.0)

plt.show()
```

```python
# Construct a DataFrame for the evaluation data

df = pd.DataFrame(acc_streaks.items(), columns=['model', 'dict',])

df['streak'] = df['dict'].map(lambda x: x['streak']).map(lambda x: round(sum(x)/len(x)))
df['accuracy'] = df['dict'].map(lambda x: x['accuracy']).map(lambda x: round(sum(x)/len(x), 3))
df['samples per symbol'] = df['dict'].map(lambda x: x['samples'])
df.drop(['dict'], inplace=True, axis=1)

df_sorted = df.sort_values('streak', ascending=False).head(10)[['model','streak', 'samples per symbol', 'accuracy']]
minim = min(df_sorted['streak'])
maxim = max(df_sorted['streak'])
print(f'The top 10 accurate streaks of evaluation range from {round(minim)} to {round(maxim)} correct consecutive predictions.')

ICD.display(df_sorted)

for days in time_range:
    days = str(days)
    print(f'Top streaks for {days} day(s) evaluations.')
    df_ = df[df['model'].map(lambda x: x.split('-')[0] == days)].sort_values('streak', ascending=False).head(5)    
    df_['t'] = df_['model'].map(lambda x: x.split('-')[0])
    df_['n1'] = df_['model'].map(lambda x: x.split('-')[1])
    df_['n2'] = df_['model'].map(lambda x: x.split('-')[2])
    ICD.display(df_[['t', 'n1', 'n2','streak', 'samples per symbol', 'accuracy']])
```

## Evaluation: only `2*t` testing samples

```python
# Evaluate each trained model on thei respective testing set, using only a number of samples equal to twice the 
# prediction period for each model.

evaluation_2t = {}
j = 0
for t in time_range:
    for n1 in periods:
        for n2 in periods:
            
            num_models = len(time_range) * len(periods) * len(periods)
            name = f'{t}-{n1}-{n2}'
            filepath = model_dir + name + '.h5'
            model = create_model(t, n1, n2)
            model.load_weights(filepath)
            model.compile(optimizer=Adam(learning_rate=lr), loss='binary_crossentropy', metrics=['accuracy'])
    
            evaluation_2t[name] = {}
            evaluation_2t[name]['evaluations'] = {}

            for symbol in ndxt_tickers:
                
                evaluation_2t[name]['evaluations'][symbol] = []
                data_X = pd.read_csv(f'{data_dir}{time_words[t]}/{n1}_{n2}/test/{symbol}.csv', header=None).drop(labels=0, axis=1)
                data_y = pd.read_csv(f'{data_dir}{time_words[t]}/{n1}_{n2}/test/{symbol}.csv', header=None)[0]
                length = len(data_X)
                
                samples = min(2*t, length)
                evaluation_2t[name]['samples'] = samples
                for i in range(samples):
                    x = data_X[i:i+1]
                    y = data_y[i:i+1]
                    evaluation_2t[name]['evaluations'][symbol].append(model.evaluate(x, y, verbose=0)[1])
            

            j += 1
            print(f'{round(j*100/num_models, 1)}%', end=' ')

with open('results/evaluation_2t.pickle', 'wb') as f:
    pickle.dump(evaluation_2t, f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Calculate average accuracies for each symbol in evaluations of every model

acc_streaks_2t = {}
for mod, evals in evaluation_2t.items():
    acc_streaks_2t[mod] = {}
    acc_streaks_2t[mod]['accuracy'] = []
    acc_streaks_2t[mod]['streak'] = []
    acc_streaks_2t[mod]['samples'] = evals['samples']
    
    
    for symbol, accuracies in evals['evaluations'].items():
        acc = round(sum(accuracies)/len(accuracies),2)
        
        streak = 0
        for i, a in enumerate(accuracies):
            if a != 1: 
                streak = i
                break
        
        acc_streaks_2t[mod]['accuracy'].append(acc)
        acc_streaks_2t[mod]['streak'].append(streak)
        
        
```
